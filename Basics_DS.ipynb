```python
#Recursive task using Recursive function call(calling the same function)
#So this needs to be handled with start & stop conditions
#in this recursion, we are using one of the argument itself as a decrement counter so that until it reaches value 0
#the function calls itself
#Recursion mostly is in local scope and hence arguments become the main factor in carrying out recursive behavior.
def fibonacci_recursion(arg1,arg2,arg3):
    next_num = arg1 + arg2
    #write a check of no.of numbers verified till now
    if(arg3 > 0):
        arg1 = arg2
        arg2 = next_num
        arg3 = arg3 - 1
        print(next_num)
        fibonacci_recursion(arg1,arg2,arg3)
    else:
        return

fibonacci_recursion(0,1,50)

#Recursive function modification using continous trace back
#Stacking each previous items on memory...that's the recursion flow.
def Count_handshakes(total_persons):
    print('function is called with value:' + str(total_persons))
    #write a check of no.of numbers verified till now
    if(total_persons <= 2):
        total_handshakes = 1;
    else:
        total_handshakes = Count_handshakes(total_persons -1) + (total_persons - 1)
    
    return total_handshakes

Count_handshakes(total_persons = 5)
```

```python
#Aggregation & Summarization
import pandas as pd
data = pd.read_csv('datasets/big_mart_sales.csv')
#print(data.head())
#Basic aggregation(Sum,mean,median,mode)
data.Item_MRP.sum()
data.Item_MRP.mean()
data.Item_MRP.median()
data.Item_MRP.mode()
data.Outlet_Type.value_counts()
data.describe() # Summary of the Data(with statistics on numerical data)

#Missing values count
data.isna().sum()

#Group-By the data(single or multiple columns)...here below used 2 columns...first by outlet size and then on item type
d = data.groupby(['Outlet_Size','Item_Type']).mean()
d #...gives the first head rows after group-by 
#d.mean(['Item_Type'])
#d.max(['Item_Type'])
#d['Item_Type']
d['Item_MRP'] # Here we are retreiving the list of values for Item MRP for the grouped data result

#Pivot table( To group data based on column and the value column aggregation)
data
pd.pivot_table(data,index=['Outlet_Size','Item_Type'],values='Item_MRP',aggfunc='mean') # This means...we are grouping data on Item Type column with value aggregation of 'mean' on Item MRP column

#Cross tab(Frequency table of 2 or more factors)
pd.crosstab(data['Outlet_Size'],data['Item_Type'])

avg_item_visibility = data.groupby(['Item_Identifier'])['Item_Visibility'].mean().reset_index()
avg_item_visibility

#using an custom function to check item visibility of each item identifier
def get_item_visibility(x):
    return avg_item_visibility.loc[(avg_item_visibility.Item_Identifier == x),'Item_Visibility'].values[0]

#call the func through an identifier
get_item_visibility('DRA12')

#Adding the new column to the original data record.
data['avg_item_visibility'] = data.Item_Identifier.apply(get_item_visibility)
data.head()

#Transform function which actually works on the grouped data and modifies/transforms based on operation("mean")
data['avg_item_visibility2'] = data.groupby(['Item_Identifier'])['Item_Visibility'].transform('mean')
data[['avg_item_visibility','avg_item_visibility2']]

```
